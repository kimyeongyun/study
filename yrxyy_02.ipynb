```python
# 이름으로 시트 찾기
worksheet = worksheets.worksheet('시트1')

type(worksheet)
```

```python
# 순서로 시트 찾기
worksheet = worksheets.get_worksheet(0)

worksheet
```

```python
test_df = worksheet.get_all_values()
test_df = pd.DataFrame(test_df)

test_df.rename(columns=test_df.iloc[0], inplace =True)
test_df.rename(columns={'' : 'id'}, inplace=True)
test_df.drop(index=test_df.index[0], inplace=True)
test_df.reset_index(drop=True, inplace=True)

test_df.head()
```

```python
sum_l=[]
for i in range(len(test_df)):
  print('-'*50)
  root_list = test_df.iloc[i].tolist()
  row_nm = root_list[0]
  row_content = sum([float(i) for i in root_list[1:]])
  print(f'{row_nm}: {row_content:.2f}')

  sum_l.append(f'{row_content:.2f}')

sum_l
```

```python
test_df['iris_sum'] = sum_l
test_df
```
